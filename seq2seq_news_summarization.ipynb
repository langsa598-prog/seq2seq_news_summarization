```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 모델 경로
model_path = "final_model"

# 토크나이저 & 모델 로드
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSeq2SeqLM.from_pretrained(model_path)

# GPU 사용 설정
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()

print("모델 로드 완료")
```

```python
def summarize(text, max_len):
    # 토크나이징
    inputs = tokenizer(
        text,
        return_tensors="pt",
        max_length=384,
        truncation=True
    ).to(device)

    # 요약 생성
    with torch.no_grad():
        summary_ids = model.generate(
            **inputs,
            max_length=max_len,
            num_beams=4
        )

    # 디코딩
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
import gradio as gr

demo = gr.Interface(
    fn=summarize,
    inputs=[
        gr.Textbox(lines=10, placeholder="여기에 기사나 문서를 입력하세요"),
        gr.Slider(50, 200, value=128, step=10, label="요약 최대 길이")
    ],
    outputs=gr.Textbox(lines=5, label="요약 결과"),
    title="한국어 뉴스 요약 모델",
    description="Fine-tuned KoBART 기반 텍스트 요약 데모"
)

demo.launch(
    debug=True,    # 에러 생기면 바로 볼 수 있게
    share=False,   # 외부 링크 생성 끄기
    inline=True    # 코랩 결과창에 UI
)
```
